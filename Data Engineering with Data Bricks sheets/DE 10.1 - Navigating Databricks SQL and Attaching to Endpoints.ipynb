# Navigating Databricks SQL and Attaching to Endpoints

* Navigate to Databricks SQL  
  * Make sure that SQL is selected from the workspace option in the sidebar (directly below the Databricks logo)
* Make sure a SQL endpoint is on and accessible
  * Navigate to SQL endpoints in the sidebar
  * If a SQL endpoint exists and has the State **`Running`**, you'll use this endpoint
  * If a SQL endpoint exists but is **`Stopped`**, click the **`Start`** button if you have this option (**NOTE**: Start the smallest endpoint you have available to you) 
  * If no endpoints exist and you have the option, click **`Create SQL Endpoint`**; name the endpoint something you'll recognize and set the cluster size to 2X-Small. Leave all other options as default.
  * If you have no way to create or attach to a SQL endpoint, you'll need to contact a workspace administrator and request access to compute resources in Databricks SQL to continue.
* Navigate to home page in Databricks SQL
  * Click the Databricks logo at the top of the side nav bar
* Locate the **Sample dashboards** and click **`Visit gallery`**
* Click **`Import`** next to the **Retail Revenue & Supply Chain** option
  * Assuming you have a SQL endpoint available, this should load a dashboard and immediately display results
  * Click **Refresh** in the top right (the underlying data has not changed, but this is the button that would be used to pick up changes)

# Updating a DBSQL Dashboard

* Use the sidebar navigator to find the **Dashboards**
  * Locate the sample dashboard you just loaded; it should be called **Retail Revenue & Supply Chain** and have your username under the **`Created By`** field. **NOTE**: the **My Dashboards** option on the right hand side can serve as a shortcut to filter out other dashboards in the workspace
  * Click on the dashboard name to view it
* View the query behind the **Shifts in Pricing Priorities** plot
  * Hover over the plot; three vertical dots should appear. Click on these
  * Select **View Query** from the menu that appears
* Review the SQL code used to populate this plot
  * Note that 3 tier namespacing is used to identify the source table; this is a preview of new functionality to be supported by Unity Catalog
  * Click **`Run`** in the top right of the screen to preview the results of the query
* Review the visualization
  * Under the query, a tab named **Table** should be selected; click **Price by Priority over Time** to switch to a preview of your plot
  * Click **Edit Visualization** at the bottom of the screen to review settings
  * Explore how changing settings impacts your visualization
  * If you wish to apply your changes, click **Save**; otherwise, click **Cancel**
* Back in the query editor, click the **Add Visualization** button to the right of the visualization name
  * Create a bar graph
  * Set the **X Column** as **`Date`**
  * Set the **Y Column** as **`Total Price`**
  * **Group by** **`Priority`**
  * Set **Stacking** to **`Stack`**
  * Leave all other settings as defaults
  * Click **Save**
* Back in the query editor, click the default name for this visualization to edit it; change the visualization name to **`Stacked Price`**
* Add the bottom of the screen, click the three vertical dots to the left of the **`Edit Visualization`** button
  * Select **Add to Dashboard** from the menu
  * Select your **`Retail Revenue & Supply Chain`** dashboard
* Navigate back to your dashboard to view this change

# Create a New Query

* Use the sidebar to navigate to **Queries**
* Click the **`Create Query`** button
* Make sure you are connected to an endpoint. In the **Schema Browser**, click on the current metastore and select **`samples`**. 
  * Select the **`tpch`** database
  * Click on the **`partsupp`** table to get a preview of the schema
  * While hovering over the **`partsupp`** table name, click the **>>** button to insert the table name into your query text
* Write your first query:
  * **`SELECT * FROM`** the **`partsupp`** table using the full name imported in the last step; click **Run** to preview results
  * Modify this query to **`GROUP BY ps_partkey`** and return the **`ps_partkey`** and **`sum(ps_availqty)`**; click **Run** to preview results
  * Update your query to alias the 2nd column to be named **`total_availqty`** and re-execute the query
* Save your query
  * Click the **Save** button next to **Run** near the top right of the screen
  * Give the query a name you'll remember
* Add the query to your dashboard
  * Click the three vertical buttons at the bottom of the screen
  * Click **Add to Dashboard**
  * Select your **`Retail Revenue & Supply Chain`** dashboard
* Navigate back to your dashboard to view this change
  * If you wish to change the organization of visualizations, click the three vertical buttons in the top right of the screen; click **Edit** in the menu that appears and you'll be able to drag and resize visualizations
